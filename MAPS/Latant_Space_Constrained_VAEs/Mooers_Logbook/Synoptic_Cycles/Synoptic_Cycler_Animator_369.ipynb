```python
def Timestep_time_gen(starter):
    if starter == 96:
        starter = 0
    
    time = str(int(starter/4))
    remainder = starter % 4
    if remainder == 0:
        decimal =":00 LST"
    if remainder == 1:
        decimal =":15 LST"
    if remainder == 2:
        decimal =":30 LST"
    if remainder == 3:
        decimal =":45 LST"
    new_time = time + decimal
    return new_time
```
